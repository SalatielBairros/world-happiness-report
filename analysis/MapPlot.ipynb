```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
dataset = pd.read_csv('../data/complete_dataset_region/complete_dataset_region.csv')

list_countries_on_dataset = dataset['country'].unique()
list_countries_on_dataset.sort()
list_countries_on_dataset_geopandas = world['name'].unique()
list_countries_on_dataset_geopandas.sort()

print(len(list_countries_on_dataset_geopandas)) # 177
print(len(list_countries_on_dataset)) # 165
```

```python
# ADICIONA A COLUNA NAME DO GEOPANDAS RELATIVO COM A COLUNA REGION DO DATASET
for label, row in dataset.iterrows():
  for item in merge_list_country_and_names:
    if (row['country'] == item['whr']):
      dataset.loc[label, 'name'] = item['geo']

# pd.set_option('display.max_rows', None)
# dataset
# world

merged_dataset = pd.merge(world, dataset, on='name')

# pd.set_option('display.max_rows', None)
# dataset
# world
# merged_dataset
```

```python
"""
Cria um gráfico de mapa.

Parâmetros:

filter: Recebe um array de filtros
label:  Nome do gráfico
is_to_show_axis:  Mostrar ou não os eixos X e Y

"""
def plot_map(
  filter,
  label,
  is_to_show_axis
):

  # Filtro
  map_to_plot = (filter)

  # Cria o mapa
  plotted_map = map_to_plot.plot(
    column='score',
    legend=True,
    figsize=(25, 15),
    # scheme="quantiles", # Ver as dependias para colocar os missing data
    legend_kwds={
      'label': label,
      'orientation': "horizontal"
    },
    # missing_kwds={
      # "color": "lightgrey",
      # "edgecolor": "red",
      # "hatch": "///",
      # "label": "Missing values",
    # },
  )

  # Aplica configurações
  if is_to_show_axis:
    plotted_map.set_axis_off()
```

```python
# SCORE NO MUNDO EM 2021
plot_map(
  filter=(
    merged_dataset[
      (merged_dataset.score>0)
      # (merged_dataset.year==2021)
    ]
  ),
  label="Score no mundo em 2021",
  is_to_show_axis=True
)
```

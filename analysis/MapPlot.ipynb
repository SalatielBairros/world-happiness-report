```python
world['score'] = random.uniform(0, 1)

# Filtro
world = (
  world[
    (world.score>0) &
    (world.name!="Antarctica")
  ]
)

# Cria o mapa
score_map_plot = world.plot(
  column='pop_est',
  legend=True,
  figsize=(18, 13),
  legend_kwds={
    'label': "Score no mundo",
    'orientation': "horizontal"
})

# Aplica configurações
score_map_plot.set_axis_off()
```

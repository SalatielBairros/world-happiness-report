```python
countries = countries_usa_database[['Country', 'Region']].dropna()
countries.columns = ['country', 'region']
countries['region'] = countries['region'].str.split('(').str[0].str.strip().str.lower().str.replace('&', 'and')
countries['country'] = countries['country'].str.split(',').str[0].str.strip().str.lower().str.replace('&', 'and').str.replace('rep.', 'republic', regex=False)
countries.set_index('country', inplace=True)
regions = countries['region']
regions
```

```python
def get_region(country, region):
    if(region in usa_regions):
        return region
    lc = country.lower()
    if(lc in usa_countries):
        return regions[lc]     
    for uc in usa_countries:
        if(lc.startswith(uc) or uc in lc.split(' ')):
            return regions[uc]
    
    return 'Other' 
```

```python
for country in dataset.itertuples():
    print(get_region(country.country, country.region))
```
